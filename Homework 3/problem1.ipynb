```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

df = pd.read_csv('spambase.data')
df.head()
```

```python
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
cm = confusion_matrix(Y_test, model.predict(X_test))
accuracy = accuracy_score(Y_test, model.predict(X_test))
error = 1 - accuracy 
precision = precision_score(Y_test, model.predict(X_test))
recall = recall_score(Y_test, model.predict(X_test))    
f1 = f1_score(Y_test, model.predict(X_test))

print('confusion_matrix:', cm)
print('accuracy:', accuracy)
print('error:', error)
print('precision:', precision)
print('recall:', recall)
print('f1:', f1)
```

```python
coefficients = pd.Series(model.coef_[0], index=X.columns)
coefficients_sorted = coefficients.sort_values(ascending=False)

print("Top Positive Features (spam indicators):")
print(coefficients_sorted.head(10))

print("\nTop Negative Features (non-spam indicators):")
print(coefficients_sorted.tail(10))
```

```python
thresholds = [0.25, 0.5, 0.75, 0.9]
y_prob = model.predict_proba(X_test)[:,1]

for T in thresholds:
    
    y_pred = (y_prob >= T).astype(int)
    
    acc = accuracy_score(Y_test, y_pred)
    prec = precision_score(Y_test, y_pred)
    rec = recall_score(Y_test, y_pred)
    
    print(f"Threshold {T}")
    print("Accuracy:", acc)
    print("Precision:", prec)
    print("Recall:", rec)
    print()
```
