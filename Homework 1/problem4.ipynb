```python
import numpy as np

A = np.random.randint(-10, 11, size = (3, 3))

B = np.random.randint(-10, 11, size = (3, 3))

C = np.random.randint(-10, 11, size = (3, 3))

print('Matrix 1: ', A)
print('Matrix 2: ', B)
print('Matrix 3: ', C)  
```

```python
A_inv = np.linalg.inv(A)
B_inv = np.linalg.inv(B)
C_inv = np.linalg.inv(C)    

print('Inverse of Matrix 1: ', A_inv)
print('Inverse of Matrix 2: ', B_inv)
print('Inverse of Matrix 3: ', C_inv)
```

```python
print('Verification of Matrix 1: ', A @ A_inv)
print('Verification of Matrix 2: ', B @ B_inv)
print('Verfication of Matrix 3: ', C @ C_inv)
```
