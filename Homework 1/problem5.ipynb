```python
import pandas as pd
import numpy as np

df = pd.read_csv('kc_house_data.csv')

df.head()
```

```python
cols_to_exclude = ['id', 'date', 'zipcode']
summary = df.drop(columns=cols_to_exclude).agg(['mean', 'min', 'max', 'var'])
print(summary)
```

```python
cols_to_exclude = ['id', 'date', 'zipcode']
corr_table = df.drop(columns=cols_to_exclude).corr()['price'].to_frame(name='correlation')
print(corr_table)
```
